```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import FuncFormatter
import seaborn as sns

# Data from the Lib.IO Crawl
ecosystems = ['NPM', 'Maven', 'PyPI', 'NuGet', 'Go', 'Composer', 'RubyGems', 'Cargo']
years = np.array([2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025])
data = np.array([
    [415000, 647000, 941000, 1280000, 1620000, 2180000, 2730000, 4090000, 4880000],  # NPM
    [128000, 108000, 198000, 185000, 378000, 454000, 517000, 588000,  666000],      # Maven
    [83200, 118000, 174000, 232000, 311000, 418000, 508000, 501000,   601000],       # PyPI
    [79000, 111000, 145000, 200000, 265000, 296000, 412000, 472000, 543000],       # NuGet
    [862000, 811000, 1200000, 1820000, 274000, 460000, 422000, 486000, 511000],    # Go
    [143000, 187000, 253000, 313000, 365000, 344000, 400000, 407000, 442000],      # Composer
    [130000, 142000, 153000, 162000, 170000, 177000, 182000, 183000,   188000],      # RubyGems
    [7900, 13100, 22600, 35700, 53500, 75200, 102000, 137000, 174000]              # Cargo
])

# Define a color palette
palette = "coolwarm"  # Change this to any preferred palette

# Create the bar plot
plt.figure(figsize=(12, 6))  # Increase figure size

# Set up the bar width and x locations
bar_width = 0.11
x = np.arange(len(years))

# Get the color palette
colors = sns.color_palette(palette, len(ecosystems))

# Create the bars with colors in a loop
for i in range(len(ecosystems)):
    plt.bar(x + i * bar_width, data[i], width=bar_width, label=ecosystems[i], color=colors[i])

# Format y-axis
def y_formatter(y, _):
    if y >= 1_000_000:
        return f"{int(y / 1_000_000)}M"
    elif y >= 1_000:
        return f"{int(y / 1_000)}K"
    else:
        return str(int(y))

plt.gca().yaxis.set_major_formatter(FuncFormatter(y_formatter))

# Labels and title
# plt.xlabel('Year', fontsize=14)
plt.ylabel('Total Packages', fontsize=16)
# plt.title('Total Packages in Ecosystems Over the Years', fontsize=16)
plt.xticks(x + bar_width * (len(ecosystems) - 1) / 2, years, fontsize=16)
plt.yticks(fontsize=14)  # Increase y-tick font size
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.grid(axis='x', linestyle='--', linewidth=0.5)
plt.legend(fontsize=16)  # Increase legend font size
plt.tight_layout()
# Save the plot as a PDF
# plt.savefig('ecosystem_packages_over_years.pdf', format='pdf')

# Show the plot
plt.show()
```
